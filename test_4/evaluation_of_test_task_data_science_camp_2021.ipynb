## All together home task eval 

````python
def check_hometask(hometask_completed_fn):
    df_to_eval= pd.read_csv(hometask_completed_fn, index_col= 0)
    assert 'price' in list (df_to_eval)

    # read original hometask to get target indices
    df_r= pd.read_csv('/content/drive/MyDrive/Father/camp2021/test_4/data_housing.csv', index_col=0)
    inices_price_null= df_r[df_r['price'].isnull()].index

    try: 
        df_to_eval.loc[inices_price_null]
    except:
        raise Exception ('Provided dataset missed some indices')             
                       
               
    if df_to_eval.loc[inices_price_null,'price'].isnull().sum()!=0:
        raise Exception ('Target price values includes null values ')

    y_pred_home= df_to_eval.loc[inices_price_null, 'price'].values
        
    # read original price values 
    df_eval= pd.read_csv('/content/drive/MyDrive/Father/camp2021/test_4/df_eval.csv', index_col=0)
    y_test= df_eval.loc[inices_price_null, 'price_orig'].values
    
    print ('r2_score =', r2_score(y_test, y_pred_home))
           
    return r2_score(y_test, y_pred_home)
    

r2_score_cand= check_hometask(fn_candidate)
r2_score_cand
````
